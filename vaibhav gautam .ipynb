```python
import numpy as np
import matplotlib.pyplot as plt
wave1=[5*np.sin(30*np.pi*t) for t in np.linspace(0,1,2000)]
wave2=[15*np.sin(3*np.pi*t) for t in np.linspace(0,1,2000)]
wave3=[20*np.sin(20*np.pi*t) for t in np.linspace(0,1,2000)]
xn = [i+j+k for i,j,k in zip(wave1,wave2,wave3)]
m=100
l=m-1/m
xnn=list(np.zeros(m-1))+list(xn)

#sum of signal and
yn = [l*(sum(xnn[k-m+1:k]))+(1-l)*xnn[k] for k in np.arange(m-1,len(xnn))]


S=np.fft.fft(xn)
plt.plot(xn)
print(xn)
plt.plot(abs(S[0]))
plt.plot(yn)
plt.show()
```
