```python
import cv2
  
originalImage = cv2.imread('C:/Users/hp.DESKTOP-UNRH99J/Desktop/lion.jpg')
grayImage = cv2.cvtColor(originalImage, cv2.COLOR_BGR2GRAY)
  
(thresh, blackAndWhiteImage) = cv2.threshold(grayImage, 127, 255, cv2.THRESH_BINARY)
 
cv2.imshow('Black white image', blackAndWhiteImage)
cv2.imshow('Original image',originalImage)
cv2.imshow('Gray image', grayImage)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
face_cascade = cv2.CascadeClassifier('C:/Users/hp.DESKTOP-UNRH99J/Desktop/haarcascade_frontalface_default.xml')
 
image = cv2.imread('C:/Users/hp.DESKTOP-UNRH99J/Desktop/cricket team.jpg')
grayImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
 
faces = face_cascade.detectMultiScale(grayImage)
 
print type(faces)
 
if len(faces) == 0:
    print "No faces found"
 
else:
    print faces
    print faces.shape
    print "Number of faces detected: " + str(faces.shape[0])
 
    for (x,y,w,h) in faces:
        cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),1)
 
    cv2.rectangle(image, ((0,image.shape[0] -25)),(270, image.shape[0]), (255,255,255), -1)
    cv2.putText(image, "Number of faces detected: " + str(faces.shape[0]), (0,image.shape[0] -10), cv2.FONT_HERSHEY_TRIPLEX, 0.5,  (0,0,0), 1)
 
    cv2.imshow('Image with faces',image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import cv2 
import matplotlib.pyplot as plt 
img_bgr= cv2.imread('C:/Users/hp.DESKTOP-UNRH99J/Desktop/lion1.jpg')

img_neg = 1 - img_bgr 


color = ('b', 'g', 'r') 
  
for i, col in enumerate(color): 
      
    histr = cv2.calcHist([img_neg], 
                         [i], None,  
                         [256], 
                         [0, 256]) 
      
    plt.plot(histr, color = col) 
    plt.xlim([0, 256]) 
      
plt.show() 
  
plt.imshow(img_neg) 
plt.imshow(img_bgr) 
plt.show() 
```
